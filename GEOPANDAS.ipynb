```python
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point
import matplotlib.pyplot as plt
import contextily as ctx

# Load sample locations
df_locations = pd.read_csv("sample_locations.csv")
geometry_locations = [Point(xy) for xy in zip(df_locations['longitude'], df_locations['latitude'])]
gdf_locations = gpd.GeoDataFrame(df_locations, geometry=geometry_locations, crs="EPSG:4326")

# Load POIs
df_pois = pd.read_csv("pois.csv")
geometry_pois = [Point(xy) for xy in zip(df_pois['longitude'], df_pois['latitude'])]
gdf_pois = gpd.GeoDataFrame(df_pois, geometry=geometry_pois, crs="EPSG:4326")
```

```python
# Create a base plot
fig, ax = plt.subplots(figsize=(12, 10))

# Plot the sample locations with 'value' column for color
gdf_locations.to_crs(epsg=3857).plot(
    ax=ax, 
    column='value', 
    cmap='viridis', 
    legend=True, 
    markersize=50, 
    edgecolor="black",
    label="Sample Locations"
)

# Plot POIs with a different marker style
gdf_pois.to_crs(epsg=3857).plot(
    ax=ax, 
    color="red", 
    markersize=100, 
    marker="*", 
    label="POIs (Landmarks/City Centers)"
)

# Add basemap
ctx.add_basemap(ax, source=ctx.providers.OpenStreetMap.Mapnik)

# Add title and legend
plt.title("Sample Locations and Points of Interest")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.legend(loc="upper right")

plt.show()
```

```python
# Label Sample Locations
for x, y, label in zip(gdf_locations.geometry.x, gdf_locations.geometry.y, gdf_locations['name']):
    ax.text(x, y, label, fontsize=8, color='blue')

# Label POIs
for x, y, label in zip(gdf_pois.geometry.x, gdf_pois.geometry.y, gdf_pois['name']):
    ax.text(x, y, label, fontsize=10, color='red')
```
