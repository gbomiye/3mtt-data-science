```python
Cricket = Cricket.drop_duplicates()
```

```python
Cricket
```

```python
Cricket = Cricket.rename(columns={'Player': 'Player_Name', 'Span': 'Playing_Span', 'Mat': 'Matches_Played', 'Inns': 'Innings_Batted', 'NO': 'Not_Outs', 'Runs': 'Runs_Scored', 'HS': 'Highest_Inns_Scored', 'Ave': 'Battling_Average', 'BF': 'Balls_Faced', 'SR': 'Battling_Strike_Rate', '100': 'Hundreds_Scored', '50': 'Fifties_Scored', '0': 'Ducks_Scored', '4s': 'Boundary_Four', '6s': 'Boundary_Sixes'})
```

```python
Cricket
```

```python
Cricket[['Player_Name','Country']] = Cricket['Player_Name'].str.split('(', n=2, expand=True)
```

```python
Cricket
```

```python
Cricket['Country'] = Cricket['Country'].str.strip(')')
```

```python
Cricket
```

```python
Cricket[['Starting_Year', 'Ending_Year']] = Cricket['Playing_Span'].str.split('-', n=2, expand=True)
```

```python
Cricket
```

```python
Cricket['Duration'] = Cricket['Ending_Year'].astype(int) - Cricket['Starting_Year'].astype(int)
```

```python
Cricket
```

```python
Cricket = Cricket.drop(columns= 'Playing_Span')
```

```python
Cricket
```

```python
Cricket.isnull().sum()
```

```python
Cricket['Balls_Faced'] = Cricket['Balls_Faced'].fillna(0)
```

```python
Cricket
```

```python
Cricket['Balls_Faced']= Cricket['Balls_Faced'].str.split('+', expand=True)[0]
```

```python
Cricket
```

```python
Cricket = Cricket.dropna(subset=['Balls_Faced'])
```

```python
Cricket
```

```python
Cricket['Balls_Faced'] = Cricket['Balls_Faced'].astype(int)
```
